```python
import cv2
import os
import glob
import pandas as pd
from tqdm import tqdm
from ultralytics import YOLO

# โหลดโมเดล YOLOv8n (nano)
model = YOLO('yolov8n.pt')

ROOT_VIDEO_DIR = "/content/drive/MyDrive/weather_forcast/"
OUTPUT_CSV = "interpreter_dataset_info.csv"

def analyze_video(video_path, padding=0.2):
    """
    ฟังก์ชันหาพิกัดกรอบล่ามและคำนวณความยาวของคลิป
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return None

    # 1. คำนวณความยาวคลิป
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # ป้องกัน error กรณีคลิปเสียแล้วอ่าน fps ไม่ได้
    if fps > 0:
        duration_sec = total_frames / fps
        duration_min = round(duration_sec / 60, 2)
    else:
        duration_sec = 0
        duration_min = 0

    # 2. หาพิกัดกรอบล่าม
    cap.set(cv2.CAP_PROP_POS_FRAMES, total_frames // 2)
    ret, frame = cap.read()
    cap.release()

    result_data = {
        "File_Name": os.path.basename(video_path),
        "Duration_Sec": round(duration_sec, 2),
        "Duration_Min": duration_min,
        "x1": None, "y1": None, "x2": None, "y2": None,
        "Status": "Not Found",
        "Path": video_path
    }

    if not ret:
        result_data["Status"] = "Read Error"
        return result_data

    h, w = frame.shape[:2]
    results = model(frame, classes=[0], verbose=False)

    best_box = None
    min_dist_to_corner = float('inf')

    if results[0].boxes:
        for box in results[0].boxes.xyxy:
            x1, y1, x2, y2 = map(int, box[:4])
            cx, cy = (x1 + x2) // 2, (y1 + y2) // 2
            box_area = (x2 - x1) * (y2 - y1)

            if cy > h * 0.5 and box_area < (h * w) * 0.4:
                dist_bl = (cx - 0)**2 + (cy - h)**2
                dist_br = (cx - w)**2 + (cy - h)**2
                dist = min(dist_bl, dist_br)

                if dist < min_dist_to_corner:
                    min_dist_to_corner = dist
                    best_box = [x1, y1, x2, y2]

    if best_box:
        x1, y1, x2, y2 = best_box
        w_box = x2 - x1
        h_box = y2 - y1

        px = int(w_box * padding)
        py = int(h_box * padding)

        result_data["x1"] = max(0, x1 - px)
        result_data["y1"] = max(0, y1 - py)
        result_data["x2"] = min(w, x2 + px)
        result_data["y2"] = min(h, y2 + py)
        result_data["Status"] = "OK"

    return result_data

def main():
    search_pattern = os.path.join(ROOT_VIDEO_DIR, "**", "*.mp4")
    video_files = glob.glob(search_pattern, recursive=True)

    print(f"กำลังเตรียมประมวลผลไฟล์วิดีโอทั้งหมด: {len(video_files)} ไฟล์")

    if not video_files:
        print("ไม่พบไฟล์ mp4 ตรวจสอบ Path อีกครั้งครับ")
        return

    data_rows = []

    for vid_path in tqdm(video_files, desc="Analyzing Videos"):
        vid_info = analyze_video(vid_path)
        if vid_info:
            data_rows.append(vid_info)

    # นำ List ของ Dictionary มาแปลงเป็นตาราง Pandas DataFrame
    df = pd.DataFrame(data_rows)

    # คำนวณสรุปเวลาทั้งหมดให้ดูด้วย
    total_minutes = df['Duration_Min'].sum()
    total_hours = total_minutes / 60

    # บันทึกเป็น CSV
    df.to_csv(OUTPUT_CSV, index=False, encoding='utf-8-sig')

    print(f"\n✅ เสร็จสิ้น! บันทึกข้อมูลลงใน {OUTPUT_CSV}")
    print(f"📊 สรุปข้อมูล Dataset:")
    print(f"   - จำนวนคลิปทั้งหมด: {len(df)} คลิป")
    print(f"   - หาพิกัดล่ามเจอ: {len(df[df['Status'] == 'OK'])} คลิป")
    print(f"   - ความยาววิดีโอรวม: {total_minutes:.2f} นาที (ประมาณ {total_hours:.2f} ชั่วโมง)")

if __name__ == "__main__":
    main()
```

```python
import cv2
import os
import json
import torch
import numpy as np
import pandas as pd
import mediapipe as mp
import matplotlib.pyplot as plt
from moviepy.editor import VideoFileClip
from transformers import pipeline
import glob

class ASRModule:
    def __init__(self, model_id="biodatlab/whisper-th-large-v3-combined"):
        print(f"[ASR] Loading model: {model_id} ...")
        self.device = "cuda:0" if torch.cuda.is_available() else "cpu"
        self.pipe = pipeline(
            "automatic-speech-recognition",
            model=model_id,
            torch_dtype=torch.float16 if self.device == "cuda:0" else torch.float32,
            device=self.device,
        )

    def extract_audio(self, video_path, temp_audio_path="temp_audio.wav"):
        try:
            video = VideoFileClip(video_path)
            video.audio.write_audiofile(temp_audio_path, verbose=False, logger=None)
            video.close()
            return temp_audio_path
        except Exception as e:
            print(f"  [ASR Error] Failed to extract audio: {e}")
            return None

    def get_timestamps(self, audio_path, delay_sec=1.5):
        print("  -> [ASR] Transcribing and extracting timestamps...")
        result = self.pipe(audio_path, return_timestamps=True)
        sentences = []
        for chunk in result["chunks"]:
            text = chunk["text"].strip()
            if not text or chunk["timestamp"][0] is None or chunk["timestamp"][1] is None:
                continue

            # ชดเชยเวลาล่ามดีเลย์
            start_time = chunk["timestamp"][0] + delay_sec
            end_time = chunk["timestamp"][1] + delay_sec
            sentences.append({
                "Text": text,
                "Start_Time": round(start_time, 2),
                "End_Time": round(end_time, 2)
            })
        return sentences

from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import mediapipe as mp

class PoseModule:
    def __init__(self):
        print("[Pose] Initializing MediaPipe Tasks API (Pose, Hand, Face)...")

        # 1. โหลด Pose Model
        pose_base = python.BaseOptions(model_asset_path='pose_landmarker.task')
        pose_opts = vision.PoseLandmarkerOptions(base_options=pose_base, output_segmentation_masks=False)
        self.pose_landmarker = vision.PoseLandmarker.create_from_options(pose_opts)

        # 2. โหลด Hand Model (ตั้งค่า num_hands=2 เพื่อให้จับมือซ้าย-ขวาพร้อมกัน)
        hand_base = python.BaseOptions(model_asset_path='hand_landmarker.task')
        hand_opts = vision.HandLandmarkerOptions(base_options=hand_base, num_hands=2)
        self.hand_landmarker = vision.HandLandmarker.create_from_options(hand_opts)

        # 3. โหลด Face Model
        face_base = python.BaseOptions(model_asset_path='face_landmarker.task')
        face_opts = vision.FaceLandmarkerOptions(base_options=face_base, num_faces=1)
        self.face_landmarker = vision.FaceLandmarker.create_from_options(face_opts)

    def process_frame(self, img_rgb):
        """ รับภาพเข้ามา ประมวลผล 3 โมเดล แล้วคืนค่าเป็น Array 1662 มิติ """
        import numpy as np

        # Tasks API บังคับให้แปลงภาพธรรมดาเป็น mp.Image ก่อน
        mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=img_rgb)

        # สั่งรันทั้ง 3 โมเดล
        pose_result = self.pose_landmarker.detect(mp_image)
        hand_result = self.hand_landmarker.detect(mp_image)
        face_result = self.face_landmarker.detect(mp_image)

        # ==========================================
        # สกัดจุดและจัด Format ให้เหมือน Holistic เดิม
        # ==========================================

        # 1. Pose (33 จุด x 4 ค่า = 132 มิติ)
        if pose_result.pose_landmarks:
            pose = np.array([[r.x, r.y, r.z, r.visibility] for r in pose_result.pose_landmarks[0]]).flatten()
        else:
            pose = np.zeros(33 * 4)

        # 2. Face (เอาแค่ 468 จุดแรก x 3 ค่า = 1404 มิติ) *API ใหม่มี 478 จุด (มีตาขาว) เราต้องตัดออกให้เท่าเดิม
        if face_result.face_landmarks:
            face = np.array([[r.x, r.y, r.z] for r in face_result.face_landmarks[0][:468]]).flatten()
        else:
            face = np.zeros(468 * 3)

        # 3. Hands (มือซ้าย-ขวา อย่างละ 21 จุด x 3 ค่า = 63 มิติ)
        lh = np.zeros(21 * 3)
        rh = np.zeros(21 * 3)

        if hand_result.hand_landmarks and hand_result.handedness:
            # วนลูปเช็คว่ามือที่เจอคือข้างซ้ายหรือขวา
            for idx, hand_landmarks in enumerate(hand_result.hand_landmarks):
                label = hand_result.handedness[idx][0].category_name # จะได้ 'Left' หรือ 'Right'
                hand_array = np.array([[r.x, r.y, r.z] for r in hand_landmarks]).flatten()

                if label == "Left":
                    lh = hand_array
                elif label == "Right":
                    rh = hand_array

        return np.concatenate([pose, face, lh, rh])

    def close(self):
        """ เคลียร์ Memory เมื่อใช้งานเสร็จ """
        self.pose_landmarker.close()
        self.hand_landmarker.close()
        self.face_landmarker.close()
```

```python
class VisualizerModule:
    @staticmethod
    def draw_keypoints(image, keypoints_1d):
        h, w, c = image.shape

        # Pose (เขียว)
        pose = keypoints_1d[0:132].reshape(33, 4)
        for p in pose:
            x, y = int(p[0]*w), int(p[1]*h)
            if p[3] > 0.5 and x > 0 and y > 0:
                cv2.circle(image, (x, y), 3, (0, 255, 0), -1)

        # Face (ขาว)
        face = keypoints_1d[132:1536].reshape(468, 3)
        for p in face:
            x, y = int(p[0]*w), int(p[1]*h)
            if x > 0 and y > 0:
                cv2.circle(image, (x, y), 1, (255, 255, 255), -1)

        # Left Hand (น้ำเงิน)
        lh = keypoints_1d[1536:1599].reshape(21, 3)
        for p in lh:
            x, y = int(p[0]*w), int(p[1]*h)
            if x > 0 and y > 0:
                cv2.circle(image, (x, y), 4, (255, 0, 0), -1)

        # Right Hand (แดง)
        rh = keypoints_1d[1599:1662].reshape(21, 3)
        for p in rh:
            x, y = int(p[0]*w), int(p[1]*h)
            if x > 0 and y > 0:
                cv2.circle(image, (x, y), 4, (0, 0, 255), -1)

        return image

    @staticmethod
    def show_frame(image_bgr, text_label, frame_index):
        img_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(8, 8))
        plt.imshow(img_rgb)
        plt.title(f"Frame: {frame_index} | ASR: {text_label}")
        plt.axis('off')
        plt.show()
        print(f"\n[Visualizer] เฟรมที่: {frame_index}")
        print(f"[Visualizer] ข้อความ ASR: {text_label}\n")
```

```python
# ==========================================
# 3. Pipeline Orchestrator (Tasks API Edition)
# ==========================================
class SLPDatasetPipeline:
    def __init__(self, bbox_path):
        self.asr = ASRModule()
        self.pose = PoseModule()

        print(f"[Pipeline] Loading BBox Data from: {bbox_path}")
        # รองรับทั้ง CSV และ JSON
        if bbox_path.endswith('.csv'):
            df = pd.read_csv(bbox_path, encoding='utf-8-sig')
            # ปรับชื่อคอลัมน์ให้ตรงกับไฟล์ CSV ของคุณ (เช่น 'video_path' หรือ 'filename')
            self.bbox_dict = {}
            for _, row in df.iterrows():
                path = row.get('video_path') or row.get('filename')
                self.bbox_dict[path] = {
                    "x1": int(row['x1']), "y1": int(row['y1']),
                    "x2": int(row['x2']), "y2": int(row['y2'])
                }
        else:
            with open(bbox_path, 'r', encoding='utf-8-sig') as f:
                self.bbox_dict = json.load(f)

    def process_and_save(self, video_path, output_dir):
        os.makedirs(output_dir, exist_ok=True)
        video_name = os.path.basename(video_path).split('.')[0]
        bbox = self.bbox_dict.get(video_path)

        if not bbox:
            print(f"  [Skip] No BBox found for: {video_name}")
            return False

        # 1. ASR - ถอดเสียง
        temp_audio = self.asr.extract_audio(video_path)
        if not temp_audio: return False
        sentences = self.asr.get_timestamps(temp_audio)
        if os.path.exists(temp_audio): os.remove(temp_audio)

        # 2. Pose - สกัด Keypoints รายเฟรม
        cap = cv2.VideoCapture(video_path)
        fps = cap.get(cv2.CAP_PROP_FPS)
        all_keypoints = []

        print(f"  -> [Pose] Processing frames...")
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret: break

            # Crop ตามพิกัดล่าม
            crop_img = frame[bbox['y1']:bbox['y2'], bbox['x1']:bbox['x2']]
            if crop_img.size == 0:
                all_keypoints.append(np.zeros(1662))
                continue

            img_rgb = cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB)
            keypoints_1d = self.pose.process_frame(img_rgb)
            all_keypoints.append(keypoints_1d)

        cap.release()

        # 3. Save Data
        npy_path = os.path.join(output_dir, f"{video_name}.npy")
        np.save(npy_path, np.array(all_keypoints))

        mapping = []
        for s in sentences:
            start_f = int(s["Start_Time"] * fps)
            end_f = int(s["End_Time"] * fps)
            if start_f < len(all_keypoints):
                mapping.append({
                    "Video": video_name, "Text": s["Text"],
                    "Start_Frame": start_f, "End_Frame": min(end_f, len(all_keypoints)-1),
                    "Npy_Path": npy_path
                })

        pd.DataFrame(mapping).to_csv(os.path.join(output_dir, f"{video_name}_mapping.csv"), index=False, encoding='utf-8-sig')
        print(f"  [Success] {video_name} processed.")
        return True

# ==========================================
# ส่วนการรัน (Execution)
# ==========================================
if __name__ == "__main__":
    # ตรวจสอบว่าชื่อไฟล์ตรงกับที่มีในเครื่อง
    PATH_BBOX = "interpreter_bboxes_yolo.json"
    PATH_OUT = "slp_dataset_ready/"
    VIDEO_DIR = "/content/drive/MyDrive/weather_forcast/" # ปรับให้ตรงกับโฟลเดอร์ใน Drive

    builder = SLPDatasetPipeline(bbox_path=PATH_BBOX)
    video_files = glob.glob(os.path.join(VIDEO_DIR, "**", "*.mp4"), recursive=True)

    if video_files:
        # ทดสอบกับไฟล์แรกก่อน
        builder.process_and_save(video_files[0], output_dir=PATH_OUT)
```
